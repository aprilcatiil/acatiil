```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (12.0, 9.0)
```

```python
data = pd.read_csv(r'C:\Users\aprilcatiil\OneDrive\Documents\School Files\2nd Semester\Machine Learning\data.csv')
X = data.iloc[:, 1]
Y = data.iloc[:, 4]
plt.xlabel('TV')
plt.ylabel('Sales')
plt.scatter(X, Y)
plt.show()
```

```python
# Building the model
m = 0
c = 0


L = 0.00001  
epochs = 50 

n = float(len(X)) 

# Performing Gradient Descent 
for i in range(epochs): 
    Y_pred = m*X + c  
    D_m = (-2/n) * sum(X * (Y - Y_pred))  
    D_c = (-2/n) * sum(Y - Y_pred)  
    m = m - L * D_m  
    c = c - L * D_c  

print (m, c)
```

```python
# Making predictions
Y_pred = m*X + c

plt.scatter(X, Y) 
plt.plot([min(X), max(X)], [min(Y_pred), max(Y_pred)], color='red')  # regression line
plt.show()
```

```python
import statsmodels.formula.api as smf
from sklearn.linear_model import LinearRegression
from sklearn import metrics
#from sklearn.cross_validation import train_test_split
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
dataxx = pd.read_csv(r'C:\Users\aprilcatiil\OneDrive\Documents\School Files\2nd Semester\Machine Learning\TV&Sales.csv', index_col=0)
dataxx
```

```python
### STATSMODELS ###

# create a fitted model
lm1 = smf.ols(formula='Sales ~ TV', data=data).fit()

# print the coefficients
lm1.params
```

```python
### SCIKIT-LEARN ###

# create X and y
feature_cols = ['TV']
X = dataxx[feature_cols]
y = dataxx.Sales

# instantiate and fit
lm2 = LinearRegression()
lm2.fit(X, y)

# print the coefficients
print(lm2.intercept_)
print(lm2.coef_)
```

```python
# Preprocessing Input data
XX = dataxx.iloc[:, 0]
YY = dataxx.iloc[:, 1]
plt.xlabel('Advertisement_TV')
plt.ylabel('Sales')
plt.scatter(XX, YY)
plt.show()
```

```python
# lm2.coef_=slope=0.04753664
# lm2.intercept_=y-intercept=7.032593549127693
Y_predx = lm2.coef_*XX +  lm2.intercept_

plt.scatter(XX, YY) 
plt.plot([min(XX), max(XX)], [min(Y_predx), max(Y_predx)], color='red')
plt.show()
```
